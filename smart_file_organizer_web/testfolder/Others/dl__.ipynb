Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, LSTM, Dense, Dropout, Conv1D, MaxPooling1D
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
import re

# Configuration
warnings.filterwarnings('ignore')
np.random.seed(42)
tf.random.set_seed(42)

# Set plot style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("="*80)
print("TRAFFIC CONGESTION PREDICTION - COMPLETE PIPELINE")
print("="*80)
print("\n✓ Libraries imported successfully")
print(f"✓ TensorFlow version: {tf.__version__}")
print(f"✓ NumPy version: {np.__version__}")
print(f"✓ Pandas version: {pd.__version__}")
```

Load Dataset

```python
print("\n" + "="*80)
print("STEP 1: DATA LOADING")
print("="*80)

df = pd.read_csv("traffic_weather_full2020.csv")

print(f"\n✓ Dataset loaded successfully")
print(f"  Shape: {df.shape}")
print(f"  Time range: {df['5 Minutes'].min()} to {df['5 Minutes'].max()}")
print(f"\nFirst 3 rows:")
print(df.head(3))
```

Data Cleaning - Column Renaming and Timestamp

```python
print("\n" + "="*80)
print("STEP 2: DATA CLEANING")
print("="*80)

# Rename problematic column
df = df.rename(columns={'(mph)': 'speed_mph'})
print("\n✓ Renamed '(mph)' to 'speed_mph'")

# Convert timestamp
df['timestamp'] = pd.to_datetime(df['5 Minutes'])
df = df.sort_values('timestamp').reset_index(drop=True)
print("✓ Converted timestamp and sorted chronologically")
```

Data Cleaning - Numeric Conversion Function


```python
def clean_numeric(x):
    """
    Remove non-numeric characters from strings
    Keeps digits, decimal point, and minus sign
    """
    if isinstance(x, str):
        x = re.sub(r'[^0-9.\-]', '', x)
        return float(x) if x != '' else np.nan
    return float(x) if pd.notna(x) else np.nan

print("\n✓ Cleaning function defined")
```

Data Cleaning - Apply Numeric Conversion

```python
# Clean temperature fields
df['Temperature'] = df['Temperature'].apply(clean_numeric)
df['Dew Point'] = df['Dew Point'].apply(clean_numeric)

print("\n✓ Cleaned temperature fields")
print(f"  Temperature range: {df['Temperature'].min():.1f} to {df['Temperature'].max():.1f}")

# Clean other numeric fields
numeric_fields = ['Wind Speed', 'Wind Gust', 'Pressure', 'Precip.', 'Humidity']
for field in numeric_fields:
    df[field] = df[field].apply(clean_numeric)

print(f"✓ Cleaned {len(numeric_fields)} additional numeric fields")
```

Data Cleaning - Handle Missing Values


```python
print("\nMissing values before cleaning:")
missing_before = df.isnull().sum()
print(missing_before[missing_before > 0])

# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Fill categorical columns with mode
categorical_cols = ['Wind', 'Condition']
for col in categorical_cols:
    if col in df.columns and df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

print(f"\n✓ Missing values after cleaning: {df.isnull().sum().sum()}")
```

Data Cleaning - Drop Redundant Columns

```python
# Drop columns that are duplicates or not needed
cols_to_drop = ['5 Minutes', 'Hour', 'Day', 'Time']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

print(f"\n✓ Dropped redundant columns")
print(f"  Final shape after cleaning: {df.shape}")
print(f"  Remaining columns: {len(df.columns)}")
```

Feature Engineering - One-Hot Encoding


```python
print("\n" + "="*80)
print("STEP 3: FEATURE ENGINEERING")
print("="*80)

df_features = df.copy()

# One-hot encode Wind direction
if 'Wind' in df_features.columns:
    wind_dummies = pd.get_dummies(df_features['Wind'], prefix='wind', drop_first=True)
    df_features = pd.concat([df_features, wind_dummies], axis=1)
    df_features = df_features.drop(columns=['Wind'])
    print(f"\n✓ Wind directions encoded: {len(wind_dummies.columns)} dummy variables")

# One-hot encode Weather condition
if 'Condition' in df_features.columns:
    condition_dummies = pd.get_dummies(df_features['Condition'], prefix='condition', drop_first=True)
    df_features = pd.concat([df_features, condition_dummies], axis=1)
    df_features = df_features.drop(columns=['Condition'])
    print(f"✓ Weather conditions encoded: {len(condition_dummies.columns)} dummy variables")
```

Feature Engineering - Time-based Features

```python
print("\n✓ Creating time-based features...")

# Extract time components
df_features['hour'] = df_features['timestamp'].dt.hour
df_features['day_of_week'] = df_features['timestamp'].dt.dayofweek

# Cyclical encoding
df_features['hour_sin'] = np.sin(2 * np.pi * df_features['hour'] / 24)
df_features['hour_cos'] = np.cos(2 * np.pi * df_features['hour'] / 24)

print("  Created: hour, day_of_week, hour_sin, hour_cos")
```

Feature Engineering - Lag Features


```python
print("\n✓ Creating lag features...")

# Skip 5-min lag to avoid data leakage for 1-hour prediction
lags = [3, 6, 12, 24]  # 15min, 30min, 1hr, 2hr

for lag in lags:
    df_features[f'flow_lag_{lag}'] = df_features['Flow'].shift(lag)
    minutes = lag * 5
    print(f"  Created flow_lag_{lag} ({minutes} minutes ago)")
```

Feature Engineering - Rolling Features


```python
print("\n✓ Creating rolling features (past-only)...")

# 30-minute rolling window
df_features['flow_roll_mean_6'] = df_features['Flow'].shift(1).rolling(6).mean()
df_features['flow_roll_std_6'] = df_features['Flow'].shift(1).rolling(6).std()

print("  Created: rolling mean and std (30-minute window)")
```

Feature Engineering - Finalize


```python
print(f"\nShape before dropping NaN: {df_features.shape}")

df_features = df_features.dropna()

print(f"Shape after dropping NaN: {df_features.shape}")
print(f"✓ Total features created: {len(df_features.columns)}")
print("✓ Feature engineering complete")
```

Sequence Creation

```python
print("\n" + "="*80)
print("STEP 4: SEQUENCE CREATION")
print("="*80)

# Configuration
INPUT_WINDOW = 72       # 6 hours of history (72 * 5 min)
OUTPUT_HORIZON = 12     # 1 hour ahead (12 * 5 min = 60 minutes)

print(f"\nSequence Configuration:")
print(f"  Input window: {INPUT_WINDOW} timesteps = {INPUT_WINDOW * 5 / 60:.1f} hours")
print(f"  Output horizon: {OUTPUT_HORIZON} timesteps = {OUTPUT_HORIZON * 5} minutes")

# Select features
feature_cols = [col for col in df_features.columns
                if col not in ['timestamp', 'Flow', 'speed_mph']]
target_col = 'Flow'

print(f"\n✓ Features: {len(feature_cols)}")
print(f"✓ Target: {target_col}")

# Build sequences
X, y = [], []
for i in range(len(df_features) - INPUT_WINDOW - OUTPUT_HORIZON):
    X.append(df_features[feature_cols].iloc[i:i+INPUT_WINDOW].values)
    y.append(df_features['Flow'].iloc[i+INPUT_WINDOW:i+INPUT_WINDOW+OUTPUT_HORIZON].values)

X = np.array(X)
y = np.array(y)

print(f"\n✓ Sequences created:")
print(f"  X shape: {X.shape} (samples, timesteps, features)")
print(f"  y shape: {y.shape} (samples, horizon)")
```

Time-based Split

```python
print("\n" + "="*80)
print("STEP 5: TIME-BASED SPLIT")
print("="*80)

# Split ratios
TRAIN_RATIO = 0.70
VAL_RATIO = 0.15
TEST_RATIO = 0.15

n_samples = len(X)
train_end = int(n_samples * TRAIN_RATIO)
val_end = int(n_samples * (TRAIN_RATIO + VAL_RATIO))

# Perform split
X_train = X[:train_end]
y_train = y[:train_end]
X_val = X[train_end:val_end]
y_val = y[train_end:val_end]
X_test = X[val_end:]
y_test = y[val_end:]

print(f"\n✓ Split complete:")
print(f"  Train: {len(X_train):,} samples (70%)")
print(f"  Val: {len(X_val):,} samples (15%)")
print(f"  Test: {len(X_test):,} samples (15%)")
```

Scaling


```python
print("\n" + "="*80)
print("STEP 6: SCALING (NO DATA LEAKAGE)")
print("="*80)

# Scale features
scaler_X = RobustScaler()
X_train_reshaped = X_train.reshape(-1, X_train.shape[2])
scaler_X.fit(X_train_reshaped)

X_train_scaled = scaler_X.transform(X_train.reshape(-1, X_train.shape[2])).reshape(X_train.shape)
X_val_scaled = scaler_X.transform(X_val.reshape(-1, X_val.shape[2])).reshape(X_val.shape)
X_test_scaled = scaler_X.transform(X_test.reshape(-1, X_test.shape[2])).reshape(X_test.shape)

# Scale target
scaler_y = RobustScaler()
scaler_y.fit(y_train)

y_train_scaled = scaler_y.transform(y_train)
y_val_scaled = scaler_y.transform(y_val)
y_test_scaled = scaler_y.transform(y_test)

print(f"\n✓ Scaling complete")
print(f"  X mean: {X_train_scaled.mean():.6f} (should be ~0)")
print(f"  X std: {X_train_scaled.std():.6f} (should be ~1)")
```

Baseline Model - Persistence

```python
print("\n" + "="*80)
print("STEP 7: BASELINE MODEL")
print("="*80)

# Find flow lag feature
try:
    flow_lag_idx = feature_cols.index('flow_lag_3')
except:
    flow_lag_features = [i for i, col in enumerate(feature_cols) if 'flow_lag' in col]
    flow_lag_idx = flow_lag_features[0]

# Get unscaled test data
X_test_unscaled = scaler_X.inverse_transform(
    X_test_scaled.reshape(-1, X_test_scaled.shape[2])
).reshape(X_test_scaled.shape)

# Predict: repeat last known value
last_known_flow = X_test_unscaled[:, -1, flow_lag_idx]
y_pred_baseline = np.tile(last_known_flow.reshape(-1, 1), (1, OUTPUT_HORIZON))

# Evaluate
mae_baseline = mean_absolute_error(y_test, y_pred_baseline)
rmse_baseline = np.sqrt(mean_squared_error(y_test, y_pred_baseline))
r2_baseline = r2_score(y_test.flatten(), y_pred_baseline.flatten())
mape_baseline = np.mean(np.abs((y_test - y_pred_baseline) / y_test)) * 100

# Step-wise MAE
step_maes_baseline = [mean_absolute_error(y_test[:, step], y_pred_baseline[:, step])
                      for step in range(OUTPUT_HORIZON)]

print(f"\n✓ Baseline (Persistence) Results:")
print(f"  MAE: {mae_baseline:.2f} vehicles")
print(f"  RMSE: {rmse_baseline:.2f} vehicles")
print(f"  R²: {r2_baseline:.4f}")
print(f"  MAPE: {mape_baseline:.2f}%")
```

Model Building Functions

```python
print("\n" + "="*80)
print("STEP 8: MODEL DEFINITIONS")
print("="*80)

def build_gru():
    """GRU model for traffic prediction"""
    model = Sequential([
        GRU(128, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(0.3),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(0.001), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

def build_lstm():
    """LSTM model for traffic prediction"""
    model = Sequential([
        LSTM(128, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(0.3),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(0.001), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

def build_cnn_gru():
    """CNN+GRU hybrid model for traffic prediction"""
    model = Sequential([
        Conv1D(64, 3, activation='relu', padding='same',
               input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        MaxPooling1D(2),
        GRU(128),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(0.001), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

def build_lstm_gru():
    """LSTM+GRU hybrid model for traffic prediction"""
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(0.2),
        GRU(128),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(0.001), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

print("\n✓ Model functions defined:")
print("  1. GRU")
print("  2. LSTM")
print("  3. CNN+GRU")
print("  4. LSTM+GRU")

# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)
```

```python
# ============================================================================
# HYPERPARAMETER TUNING SECTION (Optional - Insert after CELL 17)
# This demonstrates systematic hyperparameter optimization
# ============================================================================

# ============================================================================
# CELL 17A: Hyperparameter Tuning Setup
# ============================================================================

print("\n" + "="*80)
print("STEP 8A: HYPERPARAMETER TUNING (OPTIONAL)")
print("="*80)

print("\n⚙️ Testing different hyperparameter configurations...")
print("This may take 15-30 minutes depending on your hardware.\n")

# Hyperparameter search space
param_grid = {
    'learning_rate': [0.001, 0.0001],
    'batch_size': [32, 64],
    'dropout_rate': [0.2, 0.3],
    'hidden_units': [64, 128]
}

print("Search Space:")
print(f"  • Learning rates: {param_grid['learning_rate']}")
print(f"  • Batch sizes: {param_grid['batch_size']}")
print(f"  • Dropout rates: {param_grid['dropout_rate']}")
print(f"  • Hidden units: {param_grid['hidden_units']}")
print(f"\nTotal configurations to test: {len(param_grid['learning_rate']) * len(param_grid['batch_size']) * len(param_grid['dropout_rate']) * len(param_grid['hidden_units'])}")


# ============================================================================
# CELL 17B: Hyperparameter Tuning - Build Function
# ============================================================================

def build_model_tuning(learning_rate, dropout_rate, hidden_units):
    """
    Build model with configurable hyperparameters for tuning
    """
    model = Sequential([
        GRU(hidden_units, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(dropout_rate),
        Dense(OUTPUT_HORIZON)
    ])

    model.compile(
        optimizer=Adam(learning_rate),
        loss=tf.keras.losses.Huber(),
        metrics=['mae']
    )

    return model


# ============================================================================
# CELL 17C: Hyperparameter Tuning - Grid Search
# ============================================================================

print("\n🔍 Starting grid search...")

# Store results
tuning_results = []

# Early stopping for tuning
early_stop_tuning = EarlyStopping(
    monitor='val_loss',
    patience=3,  # Reduced patience for faster tuning
    restore_best_weights=True,
    verbose=0
)

config_num = 0
total_configs = len(param_grid['learning_rate']) * len(param_grid['batch_size']) * \
                len(param_grid['dropout_rate']) * len(param_grid['hidden_units'])

# Grid search
for lr in param_grid['learning_rate']:
    for batch_size in param_grid['batch_size']:
        for dropout in param_grid['dropout_rate']:
            for hidden in param_grid['hidden_units']:
                config_num += 1

                print(f"\n[{config_num}/{total_configs}] Testing: LR={lr}, Batch={batch_size}, Dropout={dropout}, Hidden={hidden}")

                # Build model
                model = build_model_tuning(lr, dropout, hidden)

                # Train (reduced epochs for tuning)
                history = model.fit(
                    X_train_scaled, y_train_scaled,
                    validation_data=(X_val_scaled, y_val_scaled),
                    epochs=20,  # Reduced epochs for faster tuning
                    batch_size=batch_size,
                    callbacks=[early_stop_tuning],
                    verbose=0
                )

                # Get best validation loss
                best_val_loss = min(history.history['val_loss'])
                best_val_mae = min(history.history['val_mae'])

                # Store results
                tuning_results.append({
                    'learning_rate': lr,
                    'batch_size': batch_size,
                    'dropout_rate': dropout,
                    'hidden_units': hidden,
                    'val_loss': best_val_loss,
                    'val_mae': best_val_mae
                })

                print(f"  ✓ Val Loss: {best_val_loss:.4f}, Val MAE: {best_val_mae:.4f}")

print("\n✓ Grid search complete!")


# ============================================================================
# CELL 17D: Hyperparameter Tuning - Results Analysis
# ============================================================================

print("\n" + "="*80)
print("HYPERPARAMETER TUNING RESULTS")
print("="*80)

# Create results dataframe
tuning_df = pd.DataFrame(tuning_results)
tuning_df = tuning_df.sort_values('val_mae')

print("\nTop 5 Configurations (by Validation MAE):")
print(tuning_df.head(5).to_string(index=False))

# Best configuration
best_config = tuning_df.iloc[0]

print("\n🏆 BEST CONFIGURATION:")
print(f"  • Learning Rate: {best_config['learning_rate']}")
print(f"  • Batch Size: {int(best_config['batch_size'])}")
print(f"  • Dropout Rate: {best_config['dropout_rate']}")
print(f"  • Hidden Units: {int(best_config['hidden_units'])}")
print(f"  • Validation Loss: {best_config['val_loss']:.4f}")
print(f"  • Validation MAE: {best_config['val_mae']:.4f}")

# Save tuning results
tuning_df.to_csv('hyperparameter_tuning_results.csv', index=False)
print("\n✓ Tuning results saved to: hyperparameter_tuning_results.csv")


# ============================================================================
# CELL 17E: Visualize Hyperparameter Tuning Results
# ============================================================================

print("\n📊 Creating hyperparameter tuning visualizations...")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Learning Rate Impact
lr_grouped = tuning_df.groupby('learning_rate')['val_mae'].mean()
axes[0, 0].bar(range(len(lr_grouped)), lr_grouped.values, color='steelblue', edgecolor='black')
axes[0, 0].set_xticks(range(len(lr_grouped)))
axes[0, 0].set_xticklabels([f'{lr}' for lr in lr_grouped.index])
axes[0, 0].set_xlabel('Learning Rate', fontsize=12)
axes[0, 0].set_ylabel('Average Validation MAE', fontsize=12)
axes[0, 0].set_title('Impact of Learning Rate', fontsize=13, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3, axis='y')

# 2. Batch Size Impact
batch_grouped = tuning_df.groupby('batch_size')['val_mae'].mean()
axes[0, 1].bar(range(len(batch_grouped)), batch_grouped.values, color='coral', edgecolor='black')
axes[0, 1].set_xticks(range(len(batch_grouped)))
axes[0, 1].set_xticklabels([f'{int(bs)}' for bs in batch_grouped.index])
axes[0, 1].set_xlabel('Batch Size', fontsize=12)
axes[0, 1].set_ylabel('Average Validation MAE', fontsize=12)
axes[0, 1].set_title('Impact of Batch Size', fontsize=13, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3, axis='y')

# 3. Dropout Rate Impact
dropout_grouped = tuning_df.groupby('dropout_rate')['val_mae'].mean()
axes[1, 0].bar(range(len(dropout_grouped)), dropout_grouped.values, color='mediumseagreen', edgecolor='black')
axes[1, 0].set_xticks(range(len(dropout_grouped)))
axes[1, 0].set_xticklabels([f'{dr}' for dr in dropout_grouped.index])
axes[1, 0].set_xlabel('Dropout Rate', fontsize=12)
axes[1, 0].set_ylabel('Average Validation MAE', fontsize=12)
axes[1, 0].set_title('Impact of Dropout Rate', fontsize=13, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3, axis='y')

# 4. Hidden Units Impact
hidden_grouped = tuning_df.groupby('hidden_units')['val_mae'].mean()
axes[1, 1].bar(range(len(hidden_grouped)), hidden_grouped.values, color='mediumpurple', edgecolor='black')
axes[1, 1].set_xticks(range(len(hidden_grouped)))
axes[1, 1].set_xticklabels([f'{int(hu)}' for hu in hidden_grouped.index])
axes[1, 1].set_xlabel('Hidden Units', fontsize=12)
axes[1, 1].set_ylabel('Average Validation MAE', fontsize=12)
axes[1, 1].set_title('Impact of Hidden Units', fontsize=13, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.suptitle('Hyperparameter Sensitivity Analysis', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('TUNING_hyperparameter_impact.png', dpi=300, bbox_inches='tight')
print("✓ Saved: TUNING_hyperparameter_impact.png")
plt.show()


# ============================================================================
# CELL 17F: Update Model Functions with Best Hyperparameters
# ============================================================================

print("\n⚙️ Updating model functions with optimized hyperparameters...")

# Extract best hyperparameters
BEST_LR = best_config['learning_rate']
BEST_BATCH_SIZE = int(best_config['batch_size'])
BEST_DROPOUT = best_config['dropout_rate']
BEST_HIDDEN = int(best_config['hidden_units'])

def build_gru_optimized():
    """GRU model with optimized hyperparameters from tuning"""
    model = Sequential([
        GRU(BEST_HIDDEN, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(BEST_DROPOUT),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(BEST_LR), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

def build_lstm_optimized():
    """LSTM model with optimized hyperparameters from tuning"""
    model = Sequential([
        LSTM(BEST_HIDDEN, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(BEST_DROPOUT),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(BEST_LR), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

def build_cnn_gru_optimized():
    """CNN+GRU model with optimized hyperparameters from tuning"""
    model = Sequential([
        Conv1D(64, 3, activation='relu', padding='same',
               input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        MaxPooling1D(2),
        GRU(BEST_HIDDEN),
        Dropout(BEST_DROPOUT),
        Dense(64, activation='relu'),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(BEST_LR), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

def build_lstm_gru_optimized():
    """LSTM+GRU model with optimized hyperparameters from tuning"""
    model = Sequential([
        LSTM(BEST_HIDDEN, return_sequences=True, input_shape=(INPUT_WINDOW, X_train_scaled.shape[2])),
        Dropout(BEST_DROPOUT * 0.7),  # Slightly less dropout in first layer
        GRU(BEST_HIDDEN),
        Dropout(BEST_DROPOUT),
        Dense(64, activation='relu'),
        Dense(OUTPUT_HORIZON)
    ])
    model.compile(optimizer=Adam(BEST_LR), loss=tf.keras.losses.Huber(), metrics=['mae'])
    return model

print("\n✓ Model functions updated with optimal hyperparameters:")
print(f"  • Learning Rate: {BEST_LR}")
print(f"  • Batch Size: {BEST_BATCH_SIZE}")
print(f"  • Dropout Rate: {BEST_DROPOUT}")
print(f"  • Hidden Units: {BEST_HIDDEN}")
print("\nNote: Use these optimized functions (build_xxx_optimized) instead of the original")
print("      functions when training your final models.")

# ============================================================================
# IMPORTANT NOTE:
# After running this hyperparameter tuning section, replace the original
# model building functions (build_gru, build_lstm, etc.) with the optimized
# versions (build_gru_optimized, build_lstm_optimized, etc.) in subsequent cells
# ============================================================================
```

Train GRU Model
